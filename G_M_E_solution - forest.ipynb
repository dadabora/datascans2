**Описание датасета:**
* Id - идентификационный номер квартиры 
* DistrictId - идентификационный номер района
* Rooms - количество комнат
* Square - площадь
* LifeSquare - жилая площадь
* KitchenSquare - площадь кухни
* Floor - этаж
* HouseFloor - количество этажей в доме
* HouseYear - год постройки дома
* Ecology_1, Ecology_2, Ecology_3 - экологические показатели местности
* Social_1, Social_2, Social_3 - социальные показатели местности
* Healthcare_1, Helthcare_2 - показатели местности, связанные с охраной здоровья
* Shops_1, Shops_2 - показатели, связанные с наличием магазинов, торговых центров
* Price - цена квартиры

```python
# датасет обучения
df = df.replace('A', 0)
df = df.replace('B', 1)
# датасет для прогноза
dft = dft.replace('A', 0)
dft = dft.replace('B', 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# датасет обучения
df.loc[df['LifeSquare'].isnull(), 'LifeSquare'] = df['LifeSquare'].median()
df.loc[df['Healthcare_1'].isnull(), 'Healthcare_1'] = df['Healthcare_1'].median()
# датасет для прогноза
dft.loc[dft['LifeSquare'].isnull(), 'LifeSquare'] = dft['LifeSquare'].median()
dft.loc[dft['Healthcare_1'].isnull(), 'Healthcare_1'] = dft['Healthcare_1'].median()
```

```python
dft.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Id', 'LifeSquare', 'Healthcare_1', 'Price'], axis = 'columns'),
                                                   df['Price'], test_size = 0.3, random_state=42)
```

```python
X_train
```

#### Обучим модель

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=200, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
